DATA STRUCTURES >>>>Lists, Tuples & Dictionaries

list["login", "home"] Ordered, changeable ....Test steps, URLs


tuple("chrome", "firefox") Ordered, unchangeable .........Browser configs



dict{"user": "dona"} Key-value pairs......Test data, API responses

```python
# List — you can add/remove items
pages = ["login", "home", "dashboard"]
pages.append("profile")        # adds to end
print(f"Pages: {pages}")
print(f"First page: {pages[0]}")   # indexing starts at 0!

# Tuple — fixed, can't change
browsers = ("chrome", "firefox", "safari")
print(f"Testing on: {browsers[0]}")

# Dictionary — access by key name
test_user = {
    "username": "dona",
    "password": "test123",
    "is_admin": False}

print(f"Logging in as: {test_user['username']}")
print(f"Admin? {test_user['is_admin']}")
```

```python
# API response looks like this in Python
response = {
    "status": 200,
    "message": "Login successful",
    "token": "abc123"
}
```

```python
user = {"name": "dona", "role": "admin"}

print(f'User {user["name"]} has role {user["role"]}')
```

```python
user = {"name": "dona", "role": "admin"}

print(f'User name: "{user['name']}"')
#👉 Outer string = '...'
# 👉 Inside {} → you used 'name'

# So Python thinks:
# 'User "{user[' → string ends here ❌
# remaining part becomes invalid → hence unmatched [


# Correct one:
print(f'User name: "{user["name"]}"')
```
