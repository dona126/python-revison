Conditions n loops

```python
is_logged_in = True

if is_logged_in:
    print("✅ Login successful")
else:
    print("❌ Login failed")
```

```python
status_code = 404

if status_code == 200:
    print("✅ Page loaded successfully")
elif status_code == 404:
    print("⚠️ Page not found")
elif status_code == 500:
    print("❌ Server error")
else:                                  # no condition PRESENT
    print("🔹 Unknown status code")
```

```python
# for loop
pages = ["login", "home", "dashboard"]

for page in pages:
    print(f"Testing page: {page}")
```

```python
# while loop
attempts = 0

while attempts < 3:
    print(f"Attempt {attempts + 1}")
    attempts += 1
```
