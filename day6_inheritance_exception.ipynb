Inheritance & Exception Handling

Inheritance - Child class reuses parent class code


Exception - HandlingCatches errors gracefully instead of crashing

```python
#inheritance
# Parent class
class TestCase:
    def __init__(self, name):
        self.name = name

    def start(self):
        print(f"🚀 Starting test: {self.name}")

    def finish(self):
        print(f"✅ Test done: {self.name}")


# Child class — inherits from TestCase
class LoginTest(TestCase):
    def __init__(self, name, username):
        super().__init__(name)       # calls parent __init__! 
        # ☝️ this calls Parent's __init__
        # so you don't rewrite the same code!
        self.username = username

    def run(self):
        print(f"🔐 Running login test for {self.username}")


# Using child class
test = LoginTest("Login Test", "dona")
test.start()    # from Parent ✅
test.run()      # from Child ✅
test.finish()   # from Parent ✅
```

```python
# A parent class TestCase with __init__(name) and a start method
# A child class LoginTest that inherits from TestCase with an extra username argument and a run method
# Inside run — use try/except — if username is empty raise ValueError("Username cannot be empty!") else print login success
# Create 2 objects — one valid, one with empty username
# Call start() and run() on both!

class TestCase:
    def __init__(self, name):
        self.name = name
    def start(self):
        print("Parent: start")

class LoginTest(TestCase):
    def __init__(self, name, username):
        super().__init__(name)
        self.username = username

    def run(self):
       try:
           if self.username == "":
               raise ValueError("Username cannot be empty!") 
           #raise is used to manually create (throw) an error. 
           #raise is used to explicitly throw an exception when a condition fails, 
           #and except is used to handle that exception gracefully
           print("Login success")
       except ValueError as e:
            print(f"Error: {e}")
           

test1 = LoginTest("Login Test", "dona")
test2 = LoginTest("Login Test", "")
test1.start()   
test1.run()    
test2.start()   
test2.run()      

    
```
