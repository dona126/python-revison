Functions

#A function is a reusable block of code — write once, use many times!



```python
#Basic fn
def greet():
    print("Hello, Tester!")

greet()   # calling the function
```

```python
#With arguments
def greet(name):
    print(f"Hello, {name}!")

greet("Dona")   # Output: Hello, Dona!
```

```python
#With return
def add(a, b):
    return a + b

result = add(10, 20)
print(f"Total: {result}")   # Output: Total: 30
```

```python
#With default values
def login(username, role="user"):
    print(f"Logging in as {username} — Role: {role}")

login("dona")              # uses default role
login("admin", "admin")    # overrides default
```

```python
def check_status(page, status_code=200):
    if status_code == 200:
        print(f"✅ {page} loaded successfully")
    elif status_code == 404:
        print(f"⚠️ {page} not found")
    else:
        print(f"❌ {page} failed with {status_code}")

check_status("login")           # uses default 200
check_status("signup", 404)     # overrides to 404
check_status("dashboard", 500)  # overrides to 500
```

```python
# Is called validate_login
# Takes 2 arguments — username and password
# Has a default value for password — "test123"
# Returns True if username is "dona" and password is "test123" — else returns False
# Call it twice — once with correct details, once with wrong ones
# Print the result using f-string

def validate_login( username, password="test123"):
    if username == "dona" and password == "test123":
        return True
    else: 
        return False
        

print(f"{validate_login( 'dona', 'test123')}")
print(f"{validate_login( 'dd', 'test123')}")


    
```
