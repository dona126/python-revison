```python
class TestUser:
    #Every method in a class needs self as first argument!
    #Inside a class, always use self.name not just name!
    def __init__(self, username, password):
        self.username = username
        self.password = password

    def login(self): #Every method in a class needs self as first argument!
        print(f"🔐 Logging in as {self.username}") # Inside a class, always use self.name not just name!

    def logout(self):
        print(f"👋 {self.username} logged out")


# Creating objects
user1 = TestUser("dona", "test123")
user2 = TestUser("admin", "admin123")

user1.login()    # 🔐 Logging in as dona
user2.login()    # 🔐 Logging in as admin
user1.logout()   # 👋 dona logged out
```

```python
#testing theme based example

class TestPage:
    def __init__(self, name, url, status=200):
        self.name = name
        self.url = url
        self.status = status

    def check(self):
        if self.status == 200:
            print(f"✅ {self.name} — {self.url} loaded!")
        elif self.status == 404:
            print(f"⚠️ {self.name} — page not found!")
        else:
            print(f"❌ {self.name} — failed with {self.status}")


# Creating objects
login_page = TestPage("Login", "https://app.com/login")
broken_page = TestPage("Signup", "https://app.com/signup", 404)

login_page.check()    # ✅ Login — https://app.com/login loaded!
broken_page.check()   # ⚠️ Signup — page not found!
```

```python
# Write a class called TestCase that has:

# __init__ with — name, status (default "pending")
# Method run — sets status to "passed" and prints "✅ {name} passed!"
# Method fail — sets status to "failed" and prints "❌ {name} failed!"
# Method get_status — prints "Status of {name}: {status}"

# Then:

# Create 2 objects
# Call different methods on each
# Check status of both!

class TestCase:
    def __init__(self, name, status="pending"):
        self.name = name
        self.status = status 
    def run(self):
        self.status = "passed"
        print(f"✅ {self.name} passed!")

    def fail(self):
        self.status = "failed"
        print(f"❌ {self.name} failed!")

    def get_status(self):
        print(f"Status of {self.name}: {self.status}")
        


case1 = TestCase("login", "passed")
case2 = TestCase("logout", "failed")

case1.run()
case1.get_status()
case2.fail()
case2.get_status()
    

```
